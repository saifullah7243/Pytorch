````python
 # Step-1
 import torch

 x = torch.tensor(2.0, requires_grad=True)
````

````python
# step 2
## define function
y = x**2 + 3 * x
````

````python
## Step 3
## compute with .backward()
y.backward()
````

````python
## Access
print(x.grad)
````

# Gradient Tracking in PyTorch

In PyTorch:

- `requires_grad=True` makes PyTorch track that tensor for gradients.

After training, you may use:

- `x.detach()` â†’ to get a tensor without gradient tracking.

- `with torch.no_grad():` â†’ to turn off gradient tracking during inference.
