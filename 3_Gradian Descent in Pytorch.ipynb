```python
performance  = torch.tensor(df["performance"].values, dtype=torch.float32)
years_of_experience = torch.tensor(df["years_of_experience"].values, dtype=torch.float32)
projects = torch.tensor(df["projects_completed"].values, dtype=torch.float32)
bonus = torch.tensor(df["bonus"].values, dtype=torch.float32)
```

```python
### Step_1 Initlize the weight
w1 = torch.randn(1, requires_grad=True)
w2 = torch.randn(1, requires_grad=True)
w3 = torch.randn(1, requires_grad=True)
bias = torch.randn(1, requires_grad=True)

epochs = 100000
learning_rate = 0.001

for epc in range(epochs):
    predicted_bonus = w1*performance + w2*years_of_experience + w3*projects + bias
    loss = torch.mean((predicted_bonus - bonus)**2)
    loss.backward()

    with torch.no_grad():
        w1 -= learning_rate * w1.grad
        w2 -= learning_rate * w2.grad
        w3 -= learning_rate * w3.grad
        bias -= learning_rate * bias.grad

    w1.grad.zero_()
    w2.grad.zero_()
    w3.grad.zero_()
    bias.grad.zero_()

    if epc % 10 == 0:
        print(f"Epoch: {epc} Loss: {loss.item():0.2f}")
```

```python
w1.item(), w2.item(), w3.item(), bias.item()
```

```python
## Final equatioin
final_prediction = w1.item()*performance + w2.item()*years_of_experience + w3.item()*projects + bias.item()
```
